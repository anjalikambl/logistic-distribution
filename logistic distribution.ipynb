```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import logistic

x=np.linspace(-10,10,100)
y1=logistic.pdf(x,0,1)
y2=logistic.pdf(x,0,3)
y3=logistic.pdf(x,0,0.5)
y4=logistic.pdf(x,5,1)
plt.plot(x,y1,'b-',x,y2,"r-",x,y3,'g-',x,y4,'y-')
plt.xlabel('random variable')
plt.ylabel('probability')
plt.title('logistic distribution pdf')
plt.legend(['location=0 scale=1','location=0 scale=3','location=0 scale=0.5','location=5 scale=1'])
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import logistic

x=np.linspace(-10,10,100)
y1=logistic.cdf(x,0,1)
y2=logistic.cdf(x,0,3)
y3=logistic.cdf(x,0,0.5)
y4=logistic.cdf(x,5,1)
plt.plot(x,y1,'b-',x,y2,"r-",x,y3,'g-',x,y4,'y-')
plt.xlabel('random variable')
plt.ylabel('cdf')
plt.title('logistic distribution cdf')
plt.legend(['location=0 scale=1','location=0 scale=3','location=0 scale=0.5','location=5 scale=1'])
```
